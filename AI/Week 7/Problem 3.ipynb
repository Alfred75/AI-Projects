```python
%matplotlib inline
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.model_selection import cross_val_score
```

```python
data = pd.read_csv('input3.csv', sep=',')
X = data.values[:,:2]
y = data.values[:,2]
```

```python
mpl.rcParams['figure.figsize'] = (20, 10)
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y,
                                                    test_size=0.4, random_state=0, stratify = y)
X0, X1 = X[:, 0], X[:, 1]
x_min, x_max = X0.min() - 1, X0.max() + 1
y_min, y_max = X1.min() - 1, X1.max() + 1
h= 0.02
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
```

```python
plt.close('all')
cm = plt.cm.RdBu
cm_bright = ListedColormap(['#FF0000', '#0000FF'])
plt.figure().get_size_inches()
```

```python
#list of parameters used for cross-validation
models = [('svm_linear', svm.SVC(), {'kernel': ['linear'], 'C': [0.1, 0.5, 1, 5, 10, 50, 100]}),
          ('svm_polynomial',svm.SVC(), {'kernel': ['poly'], 'C': [0.1,1,3], 'degree' : [4,5,6], 'gamma' : [0.1,0.5]}),
          ('svm_rbf',svm.SVC(), {'kernel': ['rbf'], 'C': [0.1, 0.5, 1, 5, 10, 50, 100], 'gamma': [0.1, 0.5, 1, 3, 6, 10]}),
          ('logistic', LogisticRegression(), { 'C': [0.1, 0.5, 1, 5, 10, 50, 100]}),
         ('knn', KNeighborsClassifier(), {'n_neighbors' : range(1,51), 'leaf_size' : range(5,65,5)}),
         ('decision_tree', DecisionTreeClassifier(), { 'max_depth' : range(1,51), 'min_samples_split' :  range(2,11)}),
         ("random_forest", RandomForestClassifier(), { 'max_depth' : range(1,51), 'min_samples_split' :  range(2,11)})]
```

```python
i = 1
solution = []
for model in models:
    clf = GridSearchCV(model[1], model[2], cv=5)
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    
    # print results
    print(model[0]+":")
    print(clf.best_params_)
    solution.append([model[0],clf.score(X_train, y_train),score])
    print([model[0],clf.score(X_train, y_train),score])
    print()
    
    #update graphic components
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax = plt.subplot(2,4,i)
    ax.set_title(model[0])
    ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)
    # Plot training points
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors='k')
    # Plot testing points
    ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, edgecolors='k', alpha=0.6)
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xticks(())
    ax.set_yticks(())
    ax.text(xx.max() - .3, yy.min() + .3, ('%.2f' % score).lstrip('0'),
                size=15, horizontalalignment='right')
    i +=1

plt.show()
with open("output3.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerows(solution)
```
